```python
from augmentation import DataAugmentation
import torch
import cv2
from kornia import tensor_to_image, image_to_tensor
import matplotlib.pyplot as plt
```

```python
model = DataAugmentation('extreme')

fig, axs = plt.subplots(2, 2)
for i, photo in enumerate(["../dataset/1653043549.5187616/0048.jpg", "../dataset/1653042695.4914637/0076.jpg"]):
    img = cv2.imread(photo)
    inp = torch.unsqueeze(image_to_tensor(img), 0).float()
    output = model(inp)
    axs[i,0].imshow(tensor_to_image(inp.float()/255))
    axs[i,1].imshow(tensor_to_image(output.float()/255))
    print(torch.all(inp.eq(output)))
    # print(tensor_to_image(output.float()))

```
